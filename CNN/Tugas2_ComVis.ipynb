## Menampilkan Gambar

```python
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
import numpy as np

img = mpimg.imread('WW.png')


plt.imshow(img[:,:])
plt.show()

rgb_img = img
```

### Mengubah gambar menjad jpg (RGB)

```python
rgb_img = img[:,:,:3]
```

```python
plt.imshow(rgb_img)
```

## Image Processing

### Mengubah gambar menjadi hitam putih

```python
# mengubah menjadi grayscale 
# Misalkan rgb_img adalah gambar dengan 3 channel (R, G, B)
c = np.zeros((rgb_img.shape[0], rgb_img.shape[1]))  # Array kosong untuk grayscale

for i in range(rgb_img.shape[0]):
    for j in range(rgb_img.shape[1]):
        # Mengambil nilai R, G, B sebagai angka skalar
        r = rgb_img[i, j, 0]  
        g = rgb_img[i, j, 1]  
        b = rgb_img[i, j, 2]  

        # Menghitung grayscale
        c[i, j] = (r + g + b) / 3  
        
```

```python
plt.imshow(c, cmap='gray')
```

```python
# mengubah menjadi hitam putih
treshold = 20
# menngubah menjadi skala 255
c = c * 255
for i in range(rgb_img.shape[0]):
    for j in range(rgb_img.shape[1]):
        if c[i, j]>treshold:
            c[i,j] = 1
        else:
            c[i,j] = 0
        
```

```python
plt.imshow(c, cmap='gray')
```
