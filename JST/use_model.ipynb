```python
import numpy
def load_model_from_npz(file_name='model.npz'):
    data = np.load(file_name)
    Wxh = data['Wxh']
    Whh = data['Whh']
    Why = data['Why']
    bh = data['bh']
    by = data['by']
    print("Model loaded from npz.")
    return Wxh, Whh, Why, bh, by

load_model_from_npz(file_name='model_100rb.npz')
```
